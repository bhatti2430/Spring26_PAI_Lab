```python
def place_queens(size):
    chess_board = []
    for _ in range(size):
        chess_board.append(["-"] * size)
    answers = []

    def check_position(r, c):
        for previous_row in range(r):
            if chess_board[previous_row][c] == "Q":
                return False

        row, col = r - 1, c - 1
        while row >= 0 and col >= 0:
            if chess_board[row][col] == "Q":
                return False
            row -= 1
            col -= 1

        row, col = r - 1, c + 1
        while row >= 0 and col < size:
            if chess_board[row][col] == "Q":
                return False
            row -= 1
            col += 1
        return True

  
    def solve_row(current_row):
        if current_row == size:
            saved_solution = []
            for row in chess_board:
                saved_solution.append(" ".join(row))
            answers.append(saved_solution)
            return

        for column in range(size):
            if check_position(current_row, column):
                chess_board[current_row][column] = "Q"
                solve_row(current_row + 1)
                chess_board[current_row][column] = "-"  

    solve_row(0)
    return answers

n = int(input("Enter board size (N): "))
solutions_found = place_queens(n)

print("\nNumber of possible solutions:", len(solutions_found))
print()
for sol in solutions_found:
    for line in sol:
        print(line)
    print()
```
