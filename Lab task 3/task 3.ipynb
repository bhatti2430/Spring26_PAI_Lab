## **__Question 1__**

```python
X = 0
Y = 0

capX = int(input("Enter capacity of Jug X: "))
capY = int(input("Enter capacity of Jug Y: "))
goal = int(input("Enter the target amount: "))

print(f"\nStarting State: X = {X} Y = {Y}\n")

stack = [(X, Y, [])]
visited = set()

while stack:
    X, Y, path = stack.pop()

    if (X, Y) in visited:
        continue

    visited.add((X, Y))
    path = path + [(X, Y)]

    if X == goal or Y == goal:
        print("Goal reached!")
        print(f"Final State: X = {X}, Y = {Y}")
        break

    next_moves = []

    next_moves.append((capX, Y, "Fill Jug X"))
    next_moves.append((X, capY, "Fill Jug Y"))
    next_moves.append((0, Y, "Empty Jug X"))
    next_moves.append((X, 0, "Empty Jug Y"))

    transfer = min(X, capY - Y)
    next_moves.append((X - transfer, Y + transfer, "Pour X → Y"))

    transfer = min(Y, capX - X)
    next_moves.append((X + transfer, Y - transfer, "Pour Y → X"))

    for nextX, nextY, action in next_moves:
        if (nextX, nextY) not in visited:
            print(f"Applying Rule: {action} → New State: X = {nextX}, Y = {nextY}")
            stack.append((nextX, nextY, path))
else:
    print("No solution could be found.")
```
